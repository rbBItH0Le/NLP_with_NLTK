```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
with open('moby_dick_four_chapters.txt') as f:
    strst=f.read()
    
```

```python
strst
```

```python
import spacy
nlp=spacy.load('en_core_web_sm',disable=['parser','ner','tagger'])
```

```python
nlp.max_length=1198623
```

```python
strst=[token.text.lower() for token in nlp(strst) if token.text not in '\t\n\n\n!"#$%&()\n\n*+,-./:;<=>?@[\\]^_`{|}~\n-- ']
```

```python
strst
```

```python
len(strst)
```

```python
train_len=26
```

```python
text_sequences=[]
```

```python
for i in range(train_len,len(strst)):
    seq=strst[i-train_len:i]
    text_sequences.append(seq)
```

```python
text_sequences
```

```python
import tensorflow 
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts(text_sequences)
```

```python
sequences=tokenizer.texts_to_sequences(text_sequences)
```

```python
sequences
```

```python
tokenizer.word_index
```

```python
vocabulary_size=len(tokenizer.word_counts)
```

```python
vocabulary_size
```

```python
sequences=np.array(sequences)
```

```python
sequences
```

```python
X=sequences[:,:-1]
```

```python
y=sequences[:,-1]
```

```python
y
```

```python
X.shape[1]
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y=to_categorical(y,num_classes=vocabulary_size+1)
```

```python
y
```

```python
seq_len=X.shape[1]
```

```python
seq_len
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM,Embedding
```

```python
model=Sequential()
model.add(Embedding(vocabulary_size+1,output_dim=seq_len,input_length=seq_len))
model.add(LSTM(150,return_sequences=True))
model.add(LSTM(150))
model.add(Dense(50,activation='relu'))
model.add(Dense(vocabulary_size+1,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,batch_size=128,epochs=200,verbose=1)
```

```python
seed_text=' '.join(text_sequences[5])
```

```python
seed_text
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
output_text=[]
num_gen_words=25
for i in range(num_gen_words):
    encoded_text=tokenizer.texts_to_sequences([seed_text])[0]
    print(encoded_text)
    pad_encoded=pad_sequences([encoded_text],maxlen=seq_len,truncating='pre')
    pred_word_ind=model.predict_classes(pad_encoded,verbose=0)[0]
    print(pred_word_ind)
    pre_word=tokenizer.index_word[pred_word_ind]
    seed_text=seed_text+' '+pre_word
    output_text.append(pre_word)
```
